```python
import chromadb
client = chromadb.Client()
collection = client.create_collection(name= "my_collection")
```

```python
collection.add(
    documents=[
        'Maharashtra is famous for Mumbai',
        'Telangana is famous for Hyderabad',
        'Kerala is famous for its literacy'
    ],
    ids = ['id1','id2', 'id3']
)
```

```python
all_docs = collection.get()
all_docs
```

```python
results = collection.query(
    query_texts=['High Muslims population percentage'],
    n_results = 3
)
results
```

```python
collection.delete(ids =all_docs['ids'])
collection.get()
```
